```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv('/content/drive/MyDrive/Dataset/Loan-Approval-Prediction.csv')
data.head()
```

```python
X = data.drop('Loan_Status', axis=1)
y = data['Loan_Status']
```

```python
X['TotalIncome'] = X['ApplicantIncome'] + X['CoapplicantIncome']
X['LoanAmount/TotalIncome'] = X['LoanAmount'] / X['TotalIncome']
X['Loan_Amount_Term/LoanAmount'] = X['Loan_Amount_Term'] / X['LoanAmount']
```

```python
X = X.drop(['Loan_ID', 'ApplicantIncome', 'CoapplicantIncome'], axis=1)
```

```python
imputer = SimpleImputer(strategy='most_frequent')
X_imputed = imputer.fit_transform(X)
encoder = LabelEncoder()
for i in range(X_imputed.shape[1]):
    if isinstance(X_imputed[:, i][0], str):
        X_imputed[:, i] = encoder.fit_transform(X_imputed[:, i])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
