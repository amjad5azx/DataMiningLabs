```python
import cmath
a = int(input("Insert value of a: "))
b = int(input("Insert value of b: "))
c = int(input("Insert value of c: "))
d = cmath.sqrt((b**2) - 4*a*c)
x1 = (-b + d) / (2*a)
x2 = (-b - d) / (2*a)
print("\nResult: ")
print("x1: ", x1)
print("x2: ", x2)
```
