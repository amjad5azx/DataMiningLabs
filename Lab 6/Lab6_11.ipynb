```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
def vader_sentiment_analysis(sentence):
    score = analyzer.polarity_scores(sentence)
    compound_score = score['compound']
    if compound_score > 0.05:
        return 'Positive'
    elif compound_score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
print("Vader Ruled Based Classification:")
print("I love spending time with my family.: ", vader_sentiment_analysis("I love spending time with my family."))
print("I hate when my computer crashes.: ", vader_sentiment_analysis("I hate when my computer crashes."))
print("The weather is beautiful today.: ", vader_sentiment_analysis("The weather is beautiful today."))
print("I don't know what to think about this movie.: ", vader_sentiment_analysis("I don't know what to think about this movie."))
print("I am feeling neutral about this situation.: ", vader_sentiment_analysis("I am feeling neutral about this situation."))
```

```python
from textblob import TextBlob
def textblob_sentiment_analysis(sentence):
    blob = TextBlob(sentence)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
print("TextBlob Ruled Based Classification:")
print("I love spending time with my family.: ", textblob_sentiment_analysis("I love spending time with my family."))
print("I hate when my computer crashes.: ", textblob_sentiment_analysis("I hate when my computer crashes."))
print("The weather is beautiful today.: ", textblob_sentiment_analysis("The weather is beautiful today."))
print("I don't know what to think about this movie.: ", textblob_sentiment_analysis("I don't know what to think about this movie."))
print("I am feeling neutral about this situation.: ", textblob_sentiment_analysis("I am feeling neutral about this situation."))
```
