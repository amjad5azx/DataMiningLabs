```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
dataset =pd.read_csv("/content/Mobile_price_Dataset.csv")
dataset
```

```python
#Task 1:Identify the potential features and Label that affect the performance (2 points)
features=["battery_power","dual_sim","four_g","int_memory","m_dep","n_cores","pc","px_height","px_width","ram","talk_time","touch_screen","wifi"]
X=dataset[features]
y=dataset.price_range
```

```python
#Task 2: Perform statistical analysis on the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=109)
```

```python
#Task 3:
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Task 4: By using Naive Bayes
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
from sklearn import metrics
print("Accuracy With Naive Bayes: " ,metrics.accuracy_score(y_test,y_pred))
```

```python
#Task 4: By using SVM Linear Kernal
from sklearn import svm
from sklearn.metrics import accuracy_score
linear_kernel = svm.SVC(kernel='linear')
linear_kernel.fit(X_train, y_train)
linear_kernel_y_pred = linear_kernel.predict(X_test)
print("Accuracy with Linear kernel:", accuracy_score(y_test, linear_kernel_y_pred))

#Task 4: By using SVM Poly Kernal
poly_kernel = svm.SVC(kernel='poly', degree=3)
poly_kernel.fit(X_train, y_train)
poly_kernel_y_pred = poly_kernel.predict(X_test)
print("Accuracy with Polynomial kernel:", accuracy_score(y_test, poly_kernel_y_pred))

#Task 4: By using SVM Guassian Kernal
gaussian_kernel = svm.SVC(kernel='rbf')
gaussian_kernel.fit(X_train, y_train)
gaussian_kernel_y_pred = gaussian_kernel.predict(X_test)
print("Accuracy with Gaussian kernel:", accuracy_score(y_test, gaussian_kernel_y_pred))
```
